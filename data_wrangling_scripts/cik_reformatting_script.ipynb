```python
import csv
import pandas as pd

df = pd.read_csv("C:/Users/Ryan Goh/Desktop/AAER/aaer_no_nan3.csv")
df
```

```python
#i want to get the cik as a list
cik = df['cik']
cik_list = list(cik)

final = []
for c in cik_list:
  if len(str(c)) < 10:
    #count the diff in length
    n = 10 - len(str(c))
    zero_string = n * '0'
    final_c = zero_string + str(c)
    final.append(final_c)

final

#save this into a csv
#save the file to path
outdir = 'C:/Users/Ryan Goh/Desktop/AAER'

pd.DataFrame(final).to_csv(f"{outdir}/cik.csv")

```
